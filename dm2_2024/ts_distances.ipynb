```python
import pandas as pd
from sktime.datasets import load_gunpoint
from sktime.clustering.k_means import TimeSeriesKMeans
import seaborn as sns
from sktime.utils.plotting import plot_series
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sktime.distances import distance
```

# Distances

```python
X_train, y_train = load_gunpoint(split="TRAIN", return_type="numpy2D")
X_train.shape, y_train.shape
```

```python
plt.plot(X_train[0])
plt.plot(X_train[9])
plt.show()
```

```python
print("Euclidean distance: ", distance(X_train[0], X_train[9], metric="euclidean"))
print(
    "Squared euclidean distance: ", distance(X_train[0], X_train[9], metric="squared")
)
print("Dynamic time warping distance: ", distance(X_train[0], X_train[9], metric="dtw"))
print(
    "Derivative dynamic time warping distance: ",
    distance(X_train[0], X_train[9], metric="ddtw"),
)
print(
    "Weighted dynamic time warping distance: ",
    distance(X_train[0], X_train[9], metric="wdtw"),
)
print(
    "Weighted derivative dynamic time warping distance: ",
    distance(X_train[0], X_train[9], metric="wddtw"),
)
print(
    "Longest common subsequence distance: ",
    distance(X_train[0], X_train[9], metric="lcss"),
)
print(
    "Edit distance for real sequences distance: ",
    distance(X_train[0], X_train[9], metric="edr"),
)
print(
    "Edit distance for real penalty distance: ",
    distance(X_train[0], X_train[9], metric="erp"),
)
```
